```python
import random 


class Graph():

    def __init__(self):
        self.nodes = []

    def add_nodes(self, node):
        self.nodes.append(node)

    def get_nodes(self):
        return self.nodes


class Node():

    def __init__(self, vertex):
        self.neighbour = []
        self.vertex = vertex
        self.state = None

    def add_neighbours(self, neighbours):
        self.neighbour.append(neighbours)
        
    def get_neighbours(self):
        return self.neighbour

    def get_vertex(self):
        return self.vertex

    
class QueueNode(object):
    
    def __init__(self, d):
        self.data = d
        self.next = None
        

class Queue(object):
    
    def __init__(self):
        self.top = None
        self.last = None
        
    def add(self, item):
        queue_node = QueueNode(item)
        if self.last is not None:
            self.last.next = queue_node
        self.last = queue_node
        if self.top is None:
            self.top = self.last
        
    def remove(self):
        data = self.top.data
        self.top = self.top.next
        if self.top is None:
            self.last = None
        return data
    
    def is_empty(self):
        return self.top == None
    
    def print_nodes(self):
        nodes = []
        node = self.top
        while node is not None:
            nodes.append(node.data.get_vertex())
            node = node.next
        return nodes

    
def search(g, start, end):
    print(f'{"="*20} search ')
    print(f'start : {start.get_vertex()}     end : {end.get_vertex()}')
    if start == end:
        return True
    
    # queue
    q = Queue()
    
    # Initialize nodes state
    for u in g.get_nodes():
        u.state = 'Unvisited'
        
    start.state = 'Visited'
    
    q.add(start)
    
    while not q.is_empty():
        print(q.print_nodes())
        u = q.remove() # deque
        if u is not None:
            for v in u.get_neighbours():
                if v.state == 'Unvisited':
                    if v.get_vertex() == end.get_vertex():
                        return True
                    else:
                        v.state = 'Visiting'
                        q.add(v)
            u.state = 'Visited'
    return False


graph = Graph()

temp_nodes = []

# Generate nodes
for i in range(10):
    node = Node(i)
    temp_nodes.append(node)

# Register nodes to neighbours
for node in temp_nodes:
    node.add_neighbours(temp_nodes[random.randint(0, 9)])
    node.add_neighbours(temp_nodes[random.randint(0, 9)])
    graph.add_nodes(node)

for node in graph.get_nodes():
    neighbours = []
    for neighbour in node.get_neighbours():
        neighbours.append(neighbour.get_vertex())
    print(f'node{node.get_vertex()} : {neighbours}')
        
search(graph, temp_nodes[0], temp_nodes[5])
```

```python
# 4-2

class TreeNode(object):
    
    def __init__(self, name):
        self.name = name
        self.left = None
        self.right = None
        

def create_minimal_BST(array, start, end):
    if start > end:
        return None
    
    mid = (end + start )// 2
    node = TreeNode(array[mid])
    node.left = create_minimal_BST(array, start, mid - 1)
    node.right = create_minimal_BST(array, mid + 1, end)
    return node


def print_node(node, parent_node):
    if node is None:
        return None
    print(f'{"="*20}{node.name}  parent:{parent_node}')
    
    if node.left is not None:
        print(f'left : {node.left.name}')

    if node.right is not None:
        print(f'right : {node.right.name}')
        
    print_node(node.left, node.name)
    print_node(node.right, node.name)

        
array = [i for i in range(9)]
root = create_minimal_BST(array, 0, len(array)-1)
print_node(root, None)
```

```python
# 4-3

class Node(object):
    
    def __init__(self, d):
        self.next = None
        self.data = d
        
    def append_to_tail(self, d):
        end = Node(d)
        node = self
        while node.next is not None:
            node = node.next
        node.next = end
        
        
def create_level_linked_list(root, lists, level):
    if root is None:
        return None
    if len(lists) == level:
        node = Node(root.name)
        lists.append(node)
    else:
        node = lists[level]
        node.append_to_tail(root.name)
        
    create_level_linked_list(root.left, lists, level+1)
    create_level_linked_list(root.right, lists, level+1)
    
    return lists


def print_node(head):
    node = head
    while node.next is not None:
        print(node.data)
        node = node.next
    print(node.data)
        
        
def print_linked_list(level_linked_list):
    for i, head in enumerate(level_linked_list):
        print(f'{"="*20}level {i}')
        print_node(head)

        
array = [i for i in range(9)]
root = create_minimal_BST(array, 0, len(array)-1)

level_linked_list = create_level_linked_list(root, [], 0)
print_linked_list(level_linked_list)
```

```python
# 4-4
def get_height(node):
    if node is None:
        return -1
    
    return max(get_height(node.left), get_height(node.right)) + 1

def is_balanced(node):
    if node is None:
        return True
    
    height_diff = get_height(node.right) - get_height(node.left)
    
    if abs(height_diff) > 1:
        return False
    else:
        return is_balanced(node.right) and is_balanced(node.left)

array = [i for i in range(9)]
root = create_minimal_BST(array, 0, len(array)-1)
is_balanced(root)
```

```python
# 4-5
def check_BST(node, minimum=None, maximum=None):
    if node is None:
        return True
    
    print(f'{"="*20}node{node.name}')
    print(f'max : {maximum}')
    print(f'min : {minimum}')
    
    if minimum is not None and minimum > node.name:
        return False
    
    if maximum is not None and maximum < node.name:
        return False
    
    if not check_BST(node=node.left, minimum=None, maximum=node.name) or not check_BST(node=node.right, minimum=node.name, maximum=None):
        return False
    
    return True


array = [i for i in range(9)]
array.append(-1)
root = create_minimal_BST(array, 0, len(array)-1)
check_BST(root)
```
